```python
import pandas as pd
```

# 1. pivot table

```python
data = {'Country':['India','USA','Japan','China','India','USA'],'Year':['2010','2010','2010','2015','2015','2015'],
        'Continent':['Asia','NA','Asia','Asia','Asia','NA'],'Value':[20,10,40,50,60,55]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df,index=['Year'], aggfunc='sum')
```

```python
pd.pivot_table(df,index=['Continent'], aggfunc='sum')
```

# concat

```python
people1= {'id':[1,2,3,4],
          'name':['ab','rf','kl','ed'],
          'age':[45,25,41,30]
         }
people2= {'id':[11,22,33,44],
          'name':['awb','rfg','kal','ejd'],
          'age':[35,15,41,35]
         }
people3= {'id':[111,222,333,444],
          'name':['oab','raf','klm','ecd'],
          'age':[45,25,41,30]
         }
```

```python
df1=pd.DataFrame(people1)
df2=pd.DataFrame(people2)
df3=pd.DataFrame(people3)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
people4= {'id':[1,2,3,4],
          'name':['ab','rf','kl','ed'],
          'age':[45,25,41,30]
         }
people5= {'id':[1,2,3,4],
          'name':['ab','rf','kl','ed'],
          'sales':[535,415,541,235]
         }
```
